```python
X = np.random.binomial(n=100, p=0.04, size=1000)
Y = np.random.poisson(lam=4, size=1000)
Z = np.random.binomial(n=1000, p=0.004, size=1000)
W = np.random.binomial(n=10, p=0.4, size=1000)
plt.figure()
plt.subplot(2,3,1)
plt.hist(X, bins=range(15))
plt.xlabel('X')
plt.title('X ~ Bin(100, 0.04)')
plt.subplot(2,3,2)
plt.hist(Y, bins=range(15))
plt.xlabel('Y')
plt.title('Y ~ Poiss(4)')

plt.subplot(2,3,3)
plt.hist(X, bins=range(15), alpha=0.5, label='X')
plt.hist(Y, bins=range(15), alpha=0.5, label='Y')
plt.title('Comparison of X and Y')
plt.legend()

plt.subplot(2,3,4)
plt.hist(X, bins=range(15))
plt.xlabel('Z')
plt.title('Z ~ Bin(1000, 0.004)')
plt.subplot(2,3,5)
plt.hist(W, bins=range(15))
plt.xlabel('W')
plt.title('W ~ Bin(10, 0.4)')

plt.subplot(2,3,6)
plt.hist(X, bins=range(15), alpha=0.5, label='Z')
plt.hist(Y, bins=range(15), alpha=0.5, label='W')
plt.title('Comparison of Z and W')
plt.legend()

plt.tight_layout()
```

```python
from pandas.plotting import parallel_coordinates

titanic = sns.load_dataset('titanic')
titanic['who_is'] = titanic.apply(lambda x: 1 if x['who'] == 'man' else 2 if x['who'] == 'woman' else 0, axis=1)


del titanic['who']
del titanic['sex']
del titanic['embarked']
del titanic['class']
del titanic['adult_male']
del titanic['deck']
del titanic['embark_town']
del titanic['alive']
del titanic['alone']
del titanic['parch']
del titanic['sibsp']
del titanic['fare']

data = titanic.sample(n = 30)

plt.figure(figsize=(6, 5))
parallel_coordinates(data, 'survived', colormap='gist_rainbow')
```
